9. вектор з 25-ти елементів від 0 до 9. Сортування за спаданням

```python
vector = np.random.randint(0, 10, size=25)
print('початковий вектор:\n', vector)

sorted_vector = np.sort(vector)[::-1]
print('вектор відсортований за спаданням:\n', sorted_vector)
```

10. мода, медіана, середнє, максимальре, мінімальне значення

```python
from scipy import stats

mode = stats.mode(vector)

median = np.median(vector)

mean = np.mean(vector)

max_value = np.max(vector)

min_value = np.min(vector)

print('мода', mode)
print('медіана', median)
print('середнє', mean)
print('максимальне значення', max_value)
print('мінімальне значення', min_value)
```

11. дві матриці 4х4 з числами від 1 до 10. сума, різниця, поелементий добуток, частка

```python
matrix1 = np.random.randint(1,11, size=(4,4))
matrix2 = np.random.randint(1,11, size=(4,4))

print('матриця 1 \n', matrix1)
print('\nматриця 2 \n', matrix2)
```

```python
# сума матриць
matrix_sum = matrix1 + matrix2

# різнция матриць
matrix_diff = matrix1 - matrix2

# добуток матриць
matrix_prod = matrix1 * matrix2

# частка матриць
matrix_division = np.divide(matrix1, matrix2, where =(matrix2!=0))

print('сума матриць\n', matrix_sum)
print('різнция матриць\n', matrix_diff)
print('поелементний добуток\n', matrix_prod)
print('частка матриць\n', matrix_division)
```
